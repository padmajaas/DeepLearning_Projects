```
tok = Tokenizer(char_level=False,split=' ')

tok.fit_on_texts(mcml_train['Question'])
```

```
x_train=mcml_train['Question']
y_train=multilabel_binarizer.transform(mcml_train['Tags'])
x_test=mcml_test['Question']
y_test=multilabel_binarizer.transform(mcml_test['Tags'])
```

```
### Testing with validation data:
```

```
sequences_train1 = tok.texts_to_sequences(x_train1)
```

```
vocab_len1=len(tok.index_word.keys())
```

```
vocab_len1
```

```
sequences_matrix_train1 = sequence.pad_sequences(sequences_train1,maxlen=max_len)
```

```
sequences_matrix_train1
```

```
#vocab_len1=40000   # reducing the vocab_len1 for faster processing, In general the actual value is supposed to be taken
embedding_matrix=np.zeros((vocab_len1+1,300))
words_not_available=0
for word,i in tok.word_index.items():
    embed_vector=embeding_index.get(word)
    if embed_vector is not None:
        embedding_matrix[i]=embed_vector
    else:
      words_not_available+=1
```

```
words_not_available
# if this number is too high , you should look into alternative embeddings
# its not necessary that model performance will be low if this number is high 
# that'll be the case if these words matter in the context of the response 
```

```
def RNN(embedding_matrix,embed_dim,max_len,vocab_size):
    inputs = Input(name='inputs',shape=[max_len])
    
    layer = Embedding(vocab_size+1,embed_dim,input_length=max_len,
                      mask_zero=True,weights=[embedding_matrix],trainable=False)(inputs)
    layer = LSTM(64)(layer)
    layer = Dense(256,name='FC1')(layer)
    layer = Activation('relu')(layer)
    layer = Dropout(0.5)(layer)
    layer = Dense(10,name='out_layer')(layer)
    layer = Activation('sigmoid')(layer)
    model = Model(inputs=inputs,outputs=layer)
    return model
```

```
model1 = RNN(embedding_matrix,300,max_len,vocab_len1)
model1.summary()
```

```
model1.compile(loss='binary_crossentropy',optimizer='adam',
              metrics=['categorical_accuracy'])
```

```
sequences_testcv = tok.texts_to_sequences(x_cvtest)
sequences_matrix_testcv = sequence.pad_sequences(sequences_testcv,
                                               maxlen=max_len)
```

```
from tensorflow.keras.callbacks import ModelCheckpoint,EarlyStopping

checkpointer = ModelCheckpoint('stackoverflow_tags.hdf5', verbose=1, save_best_only=True)
earlystopper = EarlyStopping(monitor='val_loss', patience=2, verbose=1)

model1.fit(sequences_matrix_train1,y_train1,batch_size=50,
          epochs=5, callbacks=[checkpointer, earlystopper], validation_data=(sequences_matrix_testcv,y_cvtest))
```
